Categorical plots are a type of data visualization used to represent categorical variables, which are variables that take on a limited, fixed number of possible values (categories).
These plots help in comparing different categories or groups, visualizing the distribution of data, or showing the relationship between categorical and continuous variables.

Here are the primary types of categorical plots:

1. Bar Plot
Description: A bar plot displays the frequency or proportion of categories by representing them as bars.
The length or height of the bar corresponds to the value of the variable.
Use Case: Useful for comparing quantities across different categories.
Example: Visualizing the number of people in different age groups.

2. Count Plot
Description: Similar to a bar plot, but specifically used to count occurrences of each category in a dataset.
Use Case: To visualize how frequently each category occurs in a dataset.
Example: Count of different car brands in a dataset.

3. Box Plot (Box-and-Whisker Plot)
Description: A box plot shows the distribution of a dataset based on five summary statistics: minimum,
first quartile (Q1), median, third quartile (Q3), and maximum. It also highlights outliers.
Use Case: To compare the distribution of a continuous variable across categories.
Example: Comparing the income distribution across different job sectors.

4. Violin Plot
Description: A combination of a box plot and a kernel density plot. It shows the distribution of the data across different categories and also
 visualizes the density of the data at different values.
Use Case: To show both the distribution and the density of a continuous variable for different categories.
Example: Distribution of test scores for different classes.

5. Strip Plot
Description: Displays individual data points along an axis, with jittering applied to avoid overlapping points.
Use Case: Useful for showing the distribution of a small dataset or when you want to observe individual data points.
Example: Height of individuals across different groups.

6. Swarm Plot
Description: An enhanced version of the strip plot, where the points are adjusted to avoid overlap completely, forming a "swarm."
Use Case: To display the distribution of data points without overlap, making individual values clearer.
Example: Showing the distribution of marks across different subjects.

7. Point Plot
Description: A point plot displays mean values of a numerical variable for each category, with error bars representing uncertainty or variability.
Use Case: Useful for comparing means across different categories, with an added indication of variability.
Example: Average sales in different regions with error bars showing variability.

8. Boxen Plot
Description: A variation of the box plot that provides more detailed information about the distribution, particularly for larger datasets.
It splits the data into more boxes at different quantiles.
Use Case: To visualize large datasets with a detailed breakdown of their distribution.
Example: Detailed breakdown of test scores across different schools.

9. Facet Grid (Subplots)
Description: A grid of plots, where each plot shows a subset of the data divided by a categorical variable.
Itâ€™s often used to visualize the interaction between two categorical variables.
Use Case: To compare trends across multiple categories simultaneously.
Example: Sales trends of different products in different regions.
Categorical plots provide a wide range of tools to understand and compare the distributions of categorical data,
helping to visualize relationships between categories and quantitative data.

```python
import seaborn as sns
```

```python
#Import Datasets
tips = sns.load_dataset('tips')
iris = sns.load_dataset('iris')
```

```python
sns.stripplot(data=tips,x='day' ,y='total_bill', jitter=False)  #Strip plot (using axes level function)
```

```python
# using Catplot (using Figure Level function)
sns.catplot(data=tips, x='day', y = 'total_bill', kind='strip')
```

```python
#Changing in Jitter Parametres
sns.catplot(data=tips, x='day', y = 'total_bill', kind='strip',jitter=1)  #Overlapping on each other at jitter 1
```

```python
sns.catplot(data=tips, x='day', y = 'total_bill', kind='strip', hue='sex')
```

```python
#SwarmPlot
sns.catplot(data=tips, x='day', y='total_bill', kind='swarm')
```

```python
sns.swarmplot(data=tips, x='day', y='total_bill')
```

```python
sns.swarmplot(data=tips, x='day', y='total_bill',hue='sex')
```
