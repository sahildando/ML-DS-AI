```python
# Distribution plots in Seaborn are essential for understanding the distribution of data.
#  They help visualize the shape, spread, and tendency of a dataset, offering insights into whether the data is normal, skewed, or multimodal.
#  Distribution plots allow us to analyze how frequently each value occurs in the dataset, giving a clearer picture of the underlying data structure.

# Key Concepts in Distribution Plots
# Distribution: This refers to how the data points are spread or distributed across the range of values.
# The goal of distribution plots is to provide an understanding of the dataset’s characteristics, such as central tendency, spread (variance),
#  and the shape of the distribution (whether it’s normal, skewed, or has multiple peaks).

# Probability Density Function (PDF): Many distribution plots show the estimated probability of
# data points falling within a particular range, which is represented by a smooth curve.

# Types of Distribution Plots in Seaborn

# Histogram (histplot)
# Kernel Density Estimation (kdeplot)
# ECDF Plot (ecdfplot)
# Joint Distribution Plot (jointplot)
# Pair Plot (pairplot)
# Let’s go over each type in detail:

# 1. Histogram (histplot)
# A histogram is one of the simplest ways to visualize the distribution of a dataset. It divides the data into bins (or intervals)
# and counts how many data points fall into each bin. It helps show the shape of the data (whether it is normal, skewed, etc.).

# Theoretical Explanation:
# Bins: The range of data is divided into intervals (called bins), and the height of each bar represents the count of observations within that interval.
# Normalization: Histograms can be normalized to show frequencies (probabilities) instead of raw counts.

# 2. Kernel Density Estimation (kdeplot)
# Kernel Density Estimation (KDE) is a smoothed estimate of the probability density function of a dataset.
#  KDE plots smooth out the noise of the histogram by drawing a continuous curve that represents the probability distribution of the data.

# Theoretical Explanation:
# KDE: A non-parametric way to estimate the probability density function (PDF) of a variable.
#  It’s like a smoothed histogram where the curve estimates the likelihood of observing a value.
# Bandwidth: This is a parameter in KDE that controls the smoothness of the curve.
#  Larger bandwidths result in a smoother curve but may lose details, while smaller bandwidths show more details but may capture noise.

#  In the KDE plot, you get a smoothed curve that estimates the probability density of the total_bill values, which helps in understanding the data’s overall distribution.

# 3. Empirical Cumulative Distribution Function (ecdfplot)
# An Empirical Cumulative Distribution Function (ECDF) plot shows the proportion or count of observations below each value in a dataset.
# It’s useful for showing the cumulative distribution of the data.

# Theoretical Explanation:
# Cumulative Distribution: The ECDF shows how the cumulative proportion of the data increases as you move through the range of the data values.
# Monotonic Nature: Since ECDF plots are cumulative, they always rise and never fall.

# In the ECDF plot, the y-axis represents the proportion of data points below a certain value of total_bill, providing insight into the overall distribution.

# 4. Joint Distribution Plot (jointplot)
# The jointplot function is used to plot the relationship between two variables and their corresponding marginal distributions
#  (histograms or KDEs along the x and y axes). It allows you to visualize how two variables are distributed together.

# Theoretical Explanation:
# Joint Distribution: Shows how two variables vary together, often using scatterplots for numeric data.
# Marginal Distributions: Each axis also shows the distribution of each individual variable (via histogram or KDE).
# Bivariate KDE: Can be added to show the density of points where data values cluster together.

# 5. Pair Plot (pairplot)
# A pairplot creates a grid of scatter plots for each pair of variables in the dataset, combined with histograms or KDE plots on the diagonals.
#  It’s especially useful when you want to explore relationships between multiple variables in one figure.

# Theoretical Explanation:
# Pairwise Relationships: A scatter plot for each combination of two variables, revealing any relationships between them.
# Diagonal: The diagonal of the plot shows the distribution of individual variables (usually using histograms or KDEs).
# Categorical Grouping: Colors can be used to represent a third variable, usually categorical, which allows us to see how the relationship differs across categories.

# Parameters Commonly Used in Distribution Plots
# bins: Controls the number of bins in the histogram.
# kde: Whether or not to include a KDE plot along with the histogram.
# shade: Fills in the area under the KDE curve.
# hue: Adds color distinction based on a categorical variable.
# kind: Specifies the type of jointplot (scatter, kde, or hex).
# diag_kind: For pairplot, controls whether the diagonal uses histograms or KDE.
# Importance of Distribution Plots
# Visualizing Data Spread: Distribution plots give a clear understanding of how data points are spread across a variable’s range.
# Identifying Outliers: By looking at the shape of the distribution, you can easily identify outliers or unusual data points.
# Understanding Data Shape: Different shapes (e.g., normal distribution, skewed distribution) tell us about the nature of the data.
# Comparing Groups: By using the hue parameter, you can compare distributions across different categorical groups.
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#Histplot
#figure level -> displot
#axes level -> histplot -> kdeplot -> rugplot
tips = sns.load_dataset('tips')
tips.head()
```

```python
# plotting univariate histogram

sns.histplot (data=tips, x='total_bill')
```

```python
sns.displot (data=tips, x='total_bill' , kind='hist')
```

```python
#Bins Parameter
sns.displot (data=tips, x='total_bill' , kind='hist', bins=20)

```

```python
# It's also possible to visualize the distribution of a categorical variable using the logic of a histogram.
#Discrete bins are automatically set for categorical variables.

tips
```

```python
sns.displot (data=tips, x='day' , kind='hist', bins=20)   #Countplot
```

```python
# Hue Parametre
sns.displot (data=tips, x='tip' , kind='hist', hue='sex')
```

```python
# Element -> step
sns.displot (data=tips, x='tip' , kind='hist', hue='sex', element='step')
```

```python
#KDEplot
#rather than using discrete bins, a KDE plot smooths the observations with a gaussian kernel, producing a continous density estimate
tips



```

```python
sns.kdeplot(data=tips, x='total_bill')
```

```python
sns.displot(data=tips, x='total_bill', kind='kde')
```

```python
#hue -> fill
sns.displot(data=tips, x='total_bill' , hue='sex', kind='kde')
```

```python
sns.displot(data=tips, x='total_bill' , hue='sex', kind='kde', fill ='true')
```

```python
#RugPlot
#Plot marginal distributions by drawing ticks along the x and y axis
# This function is intended to complement other plot by showing the location of individual obseravations in an unobtrusive way.
sns.kdeplot(data=tips, x='total_bill' )
sns.rugplot(data=tips, x='total_bill' )
```

```python
#Bivariate Histogram
#A bivariate histogram bins the data within rectangles that tile the plot
# and then shows the count of observations within each rectangle with fill colour

sns.histplot(data=tips,x='total_bill', y='tip')

```

```python
#Bivariate KDEplot
#A bivariate KDE plot smoothes the (x,y) observations with a 2D Gaussian


sns.kdeplot(data=tips,x='total_bill', y='tip')
```
