```python
# Relational Plots in Seaborn: 
# Relational plots are used to visualize relationships between variables in a dataset.
#  In Seaborn, these plots are powerful tools to understand how different data points relate to each other,
#  particularly in scenarios where you're looking at trends or distributions between numeric variables.
#   Seaborn offers high-level functions to create relational plots, specifically relplot(), which can produce both scatter plots and line plots.
#   This function is figure-level and can be used for more complex plotting, such as creating facet plots.

# 1. Relational Plotting in Seaborn
# Relational plots visualize the relationship between two continuous variables, usually represented on the x and y axes.
#  They are useful for examining how changes in one variable correspond to changes in another.

# Scatter plots show individual data points.
# Line plots show how one variable changes with respect to another, often useful for time-series data or continuous trends.

# 2. Relplot() Function Overview
# Seaborn’s relplot() is a figure-level function that creates relational plots. It uses the kind parameter to specify the type of plot:

# kind="scatter": Creates a scatter plot to visualize the relationship between two variables.
# kind="line": Creates a line plot, typically used for visualizing time-series data or trends.
# Unlike axes-level functions (like scatterplot() or lineplot()), figure-level functions such as relplot() manage the entire figure,
#  making them ideal for producing multi-plot layouts and incorporating facets (subplots based on categorical variables).
#   This is extremely useful when you want to visualize trends across different subsets of data in a single figure.

# 3. Scatter Plot in Relational Plots
# The scatter plot is the default kind in relplot(). It visualizes individual data points by plotting them at the intersection of values from two variables
#  (x and y). It's particularly useful for showing the distribution and clustering of data points.

#  4. Line Plot in Relational Plots
# When you specify kind="line" in relplot(), it produces a line plot. Line plots are used to represent how a variable changes over time
# or across another continuous variable. This is common in time-series data where trends are key.


# 5. Faceting in Relational Plots
# Faceting allows you to create small multiples (separate plots for different subsets of data).
# This is useful when you want to compare trends or relationships across different categories.
# In relplot(), the faceting is controlled using the col, row, and hue parameters:

# hue: Color-codes different categories in the same plot.
# col: Creates separate plots for each unique category along the columns.
# row: Creates separate plots for each unique category along the rows.
# col_wrap: Wraps the plots after a specified number of columns, making it easier to handle large datasets.

# 6. Customizing Facet Plots
# Seaborn provides several parameters to customize the appearance of facet plots:

# height: Adjusts the height of the individual facets.
# aspect: Controls the width-to-height ratio of each facet.
# col_wrap: Wraps facets into multiple rows after a specified number of columns.
# legend: Controls whether a legend is displayed.
# palette: Defines the color palette for the plots.

# 7. Advantages of Using Facet Plots in Relplot
# Facet plots offer several advantages, especially when working with complex datasets:

# Visual Comparison: Faceting allows you to compare relationships across different subsets of data easily.
# Organization: By splitting the plots based on categorical variables, facet plots help organize large amounts of information.
# Color-Coding: The use of the hue parameter enables easy visual distinction between different groups or categories.


import seaborn as sns
import matplotlib.pyplot as plt
import plotly as px
```

```python
# Scatter Plot
# Scatterplot can be used with several semantic groupings which can help to understand well in a graph.
# They can plot two-dimensional graphics that can be enhanced by mapping up to three additional variables while using the
# semantics of hue, size, and style parameters. All the parameter control visual semantic which are used to identify the different subsets.
#  Using redundant semantics can be helpful for making graphics more accessible.


# Parameters:

# datapandas.DataFrame, numpy.ndarray, mapping, or sequence
# Input data structure. Either a long-form collection of vectors that can be assigned to named variables or a wide-form dataset that will be internally reshaped.

# x, y vectors or keys in data
# Variables that specify positions on the x and y axes.

# hue vector or key in data
# Grouping variable that will produce points with different colors. Can be either categorical or numeric, although color mapping will behave differently in latter case.

# size vector or key in data
# Grouping variable that will produce points with different sizes. Can be either categorical or numeric, although size mapping will behave differently in latter case.

# style vector or key in data
# Grouping variable that will produce points with different markers. Can have a numeric dtype but will always be treated as categorical.

# palettestring, list, dict, or matplotlib.colors.Colormap
# Method for choosing the colors to use when mapping the hue semantic. String values are passed to color_palette(). List or dict values imply categorical mapping, while a colormap object implies numeric mapping.

# hue_ordervector of strings
# Specify the order of processing and plotting for categorical levels of the hue semantic.

# hue_normtuple or matplotlib.colors.Normalize
# Either a pair of values that set the normalization range in data units or an object that will map from data units into a [0, 1] interval. Usage implies numeric mapping.

# sizeslist, dict, or tuple
# An object that determines how sizes are chosen when size is used. List or dict arguments should provide a size for each unique data value, which forces a categorical interpretation. The argument may also be a min, max tuple.

# size_orderlist
# Specified order for appearance of the size variable levels, otherwise they are determined from the data. Not relevant when the size variable is numeric.

# size_normtuple or Normalize object
# Normalization in data units for scaling plot objects when the size variable is numeric.

# markersboolean, list, or dictionary
# Object determining how to draw the markers for different levels of the style variable. Setting to True will use default markers, or you can pass a list of markers or a dictionary mapping levels of the style variable to markers. Setting to False will draw marker-less lines. Markers are specified as in matplotlib.

# style_orderlist
# Specified order for appearance of the style variable levels otherwise they are determined from the data. Not relevant when the style variable is numeric.

# legend“auto”, “brief”, “full”, or False
# How to draw the legend. If “brief”, numeric hue and size variables will be represented with a sample of evenly spaced values. If “full”, every group will get an entry in the legend. If “auto”, choose between brief or full representation based on number of levels. If False, no legend data is added and no legend is drawn.

# axmatplotlib.axes.Axes
# Pre-existing axes for the plot. Otherwise, call matplotlib.pyplot.gca() internally.

# kwargskey, value mappings
# Other keyword arguments are passed down to matplotlib.axes.Axes.scatter()

# Returns:
# matplotlib.axes.Axes
# The matplotlib axes containing the plot.
```

```python
gap = px.data.gapminder()
temp_df = gap [gap['country'] == 'India']
temp_df
```

```python
#axes level function

sns.lineplot(data=temp_df, x="year", y ="lifeExp")

# In Seaborn, lineplot() is an axes-level function. This means that it creates a plot directly on an existing matplotlib Axes.
#  It is used to visualize continuous data using a line graph, which is commonly used for time-series or trend data.
#  It does not create its own figure but operates on an existing one.

# Key Concepts of Axes-Level Function in lineplot()

# Direct Plotting on Axes: Since it's an axes-level function, you can combine it with other axes-level
# plots and have more control over how the plot is integrated with other plots in a figure.

# Customization and Control: You can control the size, layout, and other properties of the figure because it is directly tied to Axes.
#  For instance, you can use plt.subplot() or plt.subplots() from Matplotlib to manage multiple plots.

# Integration with Matplotlib: Since lineplot() is an axes-level function, you can customize the plot further using Matplotlib functionalities,
#  such as setting labels, titles, or adjusting tick parameters.
```

```python
#Using relplot
# In Seaborn, relplot() is a figure-level function that can create different types of relational plots,
#  including scatter plots and line plots. When using relplot() with the kind="line" argument, it functions similarly to lineplot() but provides additional functionality,
#   such as faceting, where you can create multiple subplots (small multiples) in a single figure.

# Key Concepts of relplot() in Line Plot Mode
# Figure-Level Function: Unlike lineplot(), which is an axes-level function, relplot() creates a full Figure and supports faceting across multiple subplots.

# Faceting: relplot() can easily split the data into multiple plots based on one or more categorical variables using the col and row arguments.
# This allows for creating small multiples, which is useful when comparing trends across different categories.

# Control Over Plot Layout: Since it is a figure-level function, relplot() automatically handles the layout of multiple plots.
#  It generates a grid of subplots, with each subplot showing a different subset of the data.

# Versatile Plotting: Although lineplot() is used for single plots, relplot() can produce both scatter and line plots using the kind argument.

sns.relplot(data=temp_df,  x="year", y ="lifeExp",  kind="line")
```

```python
#Facet Plot

# In Seaborn, a facet plot allows you to visualize data by splitting it across multiple subplots (called facets) based on one or more categorical variables.
#  This is useful when you want to compare the same type of plot across different subsets of the data.
#  Seaborn provides several ways to create facet plots, with the most common being:

# FacetGrid: The most flexible way to create facet plots, allowing you to map data to subplots using rows, columns, and even hue.
# relplot(): A figure-level function that allows for faceting and supports relational plots like scatter or line plots.
# catplot(): Another figure-level function that supports faceting for categorical plots (e.g., boxplots, bar plots).
# pairplot(): Specially designed for pairwise relationships, faceting the data by plotting all variables against each other.
```

```python
sns.relplot(data=tips, x='total_bill', y="tip",kind="scatter", hue='sex')  #Facet Plot
```
