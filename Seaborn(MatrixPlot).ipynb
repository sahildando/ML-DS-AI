```python
# Matrix plots are a powerful way to visualize data that is structured in a grid-like format, such as correlation matrices,
# confusion matrices, and heatmaps. In Seaborn, matrix plots are typically used to represent relationships between multiple variables simultaneously,
#  often in a tabular form, making it easy to identify patterns, clusters, and trends in high-dimensional data.

# Matrix plots are especially useful when working with datasets where variables are interrelated,
#  as they can visually summarize complex relationships between multiple variables, helping you quickly spot correlations, trends, and outliers.

# Key Matrix Plot Functions in Seaborn
# Seaborn offers several types of matrix plots, each designed for specific use cases. The most commonly used matrix plots are:

# Heatmap (heatmap)
# Clustermap (clustermap)
# Let's go through each of these in detail.

# 1. Heatmap (heatmap)
# A heatmap is a graphical representation of data where individual values are represented as colors.
# t is often used to visualize correlation matrices or frequency data, where color intensity represents the magnitude of the value.
#  Heatmaps are one of the most common ways to visualize complex relationships between numerical variables in a dataset.

# Theoretical Explanation:
# Color Mapping: In a heatmap, data values are mapped to colors. The more intense the color, the higher the value, while lighter colors represent lower values.
# Matrix Format: Heatmaps are ideal for visualizing data in a matrix format where each cell of the matrix holds a numerical value.
# Correlation Matrix: One common use of heatmaps is to represent correlation matrices, which show the relationship between multiple variables.
#  Positive correlations are shown in one color (e.g., blue), while negative correlations are shown in another (e.g., red).
# Example Use Cases:
# Visualizing correlations between multiple variables in a dataset.
# Representing confusion matrices in classification problems.
# Displaying the frequency of events in a grid.

# Parameters for Customization:
# cmap: The color palette for the heatmap (e.g., 'coolwarm', 'viridis').
# annot: Whether or not to display the data values in the cells.
# linewidths: Adds lines between cells for clarity.
# vmin and vmax: Control the range of values displayed in the heatmap, helping to normalize the color mapping.
# center: Adjusts the midpoint for the colormap, useful for emphasizing deviations from a central value (like 0 in correlation matrices).
# 2. Clustermap (clustermap)
# A clustermap is an extension of the heatmap that includes hierarchical clustering on both rows and columns.
#  It arranges the variables or observations based on similarity, so similar variables are grouped together, making it easier to interpret complex relationships.

# Theoretical Explanation:
# Clustering: Clustermaps apply hierarchical clustering algorithms to group similar rows and columns together based on their similarity. This makes patterns in the data more apparent.
# Dendrogram: Clustermaps include dendrograms that display the hierarchical structure of the clusters. This shows how different variables (or data points) are related to one another.
# Reordering: By clustering, the data is reordered in the heatmap to place similar variables next to each other, enhancing the interpretability of the data.
# Use Cases:
# Gene expression analysis: Grouping genes with similar expression profiles.
# Customer segmentation: Grouping customers with similar behaviors in a marketing dataset.
# Heatmaps with large datasets: Clustermaps can handle larger datasets and reveal more complex patterns than simple heatmaps.

# The clustermap shows how different observations (rows) and variables (columns) are clustered based on their similarity.
#  The dendrograms on the sides help visualize how the rows and columns are grouped hierarchically.

# Parameters for Customization:
# method: Specifies the linkage method for clustering (e.g., 'single', 'complete', 'average').
# metric: The distance metric used to compute similarity between observations (e.g., 'euclidean', 'manhattan').
# z_score or standard_scale: Standardizes or normalizes the data along rows or columns.
# col_cluster and row_cluster: Control whether clustering is performed on columns, rows, or both.
# Other Related Matrix Plot Types
# Although heatmap and clustermap are the most commonly used matrix plots, other matrix-related functions in Seaborn include:

# 3. Pairwise Plot (pairplot)
# While not strictly a matrix plot, pairplot creates a grid of scatter plots for each pair of variables in a dataset,
# with histograms or KDE plots on the diagonals. This is useful for understanding the pairwise relationships between multiple variables.

# Practical Applications of Matrix Plots
# Correlation Heatmaps: Understanding relationships between multiple variables, such as correlations between features in machine learning.
# Confusion Matrices: Visualizing performance in classification models, where each cell represents the count of true vs. predicted classes.
# Time Series Data: Visualizing time-dependent patterns, such as airline traffic or stock prices over time.
# Advantages of Matrix Plots
# Intuitive Representation: Matrix plots like heatmaps make it easy to understand large amounts of data by visually representing values through color intensity.
# Pattern Recognition: Clustermaps enhance pattern recognition by reordering data based on similarity, making complex relationships easier to interpret.
# Comparison Across Categories: By using color maps and clustering, matrix plots help identify areas where variables behave similarly or differently.
import seaborn as sns
import matplotlib.pyplot as plt
import plotly as px


```

```python
#Heatmap
#Plot rectengular data as a color-encoded matrix
gap = px.data.gapminder()
gap
```

```python
gap.pivot(index='country',columns='year',values='lifeExp')
```

```python
temp_df=gap.pivot(index='country',columns='year',values='lifeExp')

#Axes Level function
sns.heatmap(temp_df)

```

```python
plt.figure(figsize=(15,15))
sns.heatmap(temp_df)
```

```python
temp_df=gap[gap['continent'] =='Europe'].pivot(index='country',columns='year',values='lifeExp')
```

```python
temp_df=gap[gap['continent'] =='Europe'].pivot(index='country',columns='year',values='lifeExp')
plt.figure(figsize=(15,20))
sns.heatmap(temp_df)
```

```python
temp_df=gap[gap['continent'] =='Europe'].pivot(index='country',columns='year',values='lifeExp')
plt.figure(figsize=(15,15))
sns.heatmap(temp_df, annot=True)
```

```python
temp_df=gap[gap['continent'] =='Europe'].pivot(index='country',columns='year',values='lifeExp')
plt.figure(figsize=(15,15))
sns.heatmap(temp_df, annot=True,linewidth=0.5)
```

```python
temp_df=gap[gap['continent'] =='Europe'].pivot(index='country',columns='year',values='lifeExp')
plt.figure(figsize=(15,15)) #Linewidth
sns.heatmap(temp_df, annot=True,linewidth=0.5, cmap='viridis') #Colour
```
